```python
import arcpy

# Define your polyline feature class
input_fc = r"C:\Users\rifra\Xml-to-layer-arc-tool\HW_resources\IGNACIO_MARE_ISLAND_2_115KV_TRUE_Shapes.gdb\Sections"

# Create a search cursor to iterate through the polyline features
with arcpy.da.SearchCursor(input_fc, ["SHAPE@"]) as cursor:
    for row in cursor:
        polyline = row[0]  # Get the geometry of the polyline
        for part in polyline:
            for pnt in part:
                # Print the x and y coordinates of each vertex
                print("X: {}, Y: {}".format(pnt.X, pnt.Y))
```
